# Analyse Simulated Annealing cooling scheduler

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

$T_{k+1} = T_k * d$

$T_n = T_0 * d^{n}$

$d = (T_n/T_0)^{1/n}$

```python
maxiter = 450
point = (0.5, 0.005)

f = (point[1]/point[0]) ** (1/(maxiter))

print(f'Initial T: {point[0]}')
print(f'Decay factor: {f}')

plt.figure()
it = np.arange(0,501)
fac = [point[0]*f**i for i in it]
plt.plot( it, fac )
plt.xlabel('iteration')
plt.ylabel('T')
plt.grid()
plt.show()
```
